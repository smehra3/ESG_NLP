# Data Prep

```python
#read_10q_df = pd.read_csv("gs://esgbert_resourses/raw_data/final_read_10q_df.csv")
read_10q_df = pd.read_csv("gs://esgbert_resourses/raw_data/final_subset_read_10q_df.csv")
```

```python
!gcloud config set project w266-313108
```

```python
#subset_read_10q_df = read_10q_df[(read_10q_df.environment_score_quarter_change_pure!=0)]#&(read_10q_df.quarter==1)&(read_10q_df.year==2017)
subset_read_10q_df = read_10q_df.copy()
total_ds = subset_read_10q_df.copy()
print("length of total_ds", len(total_ds))
total_ds
```

```python
#total_ds['class_environment_score_quarter_change_pure'] = np.where(total_ds['environment_score_quarter_change_pure']==0,0,1)
total_ds['class_environment_score_quarter_change_pure'] = np.where(total_ds['environment_score_quarter_change_pure']<0,0,1)
```

```python
total_ds['Reports'] = total_ds['Reports'].astype(str)
```

```python
train_size = round(len(total_ds)*.7)
valid_size = round(len(total_ds)*.2)
test_size = len(total_ds) - round(len(total_ds)*.7) - round(len(total_ds)*.2)
print('train size:', train_size)
print('valid size:', valid_size)
print('test size:', test_size)
```

```python
train_data = total_ds[:train_size]
valid_data = total_ds[train_size:train_size+valid_size]
test_data = total_ds[train_size+valid_size:]
```

```python
print('train size:', len(train_data))
print('valid size:', len(valid_data))
print('test size:', len(test_data))
```

### Model

```python
# Build the model
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
# Train the model using the training data
model.fit(train_data.Reports, train_data.class_environment_score_quarter_change_pure)
```

```python
# Predict the categories of the train data
predicted_categories_train = model.predict(train_data.Reports)
# Predict the categories of the valid data
predicted_categories_valid = model.predict(valid_data.Reports)
# Predict the categories of the test data
predicted_categories_test = model.predict(test_data.Reports)
```

```python
# Calculate accuracy
mat = confusion_matrix(train_data.class_environment_score_quarter_change_pure, predicted_categories_train)
print("The train accuracy is {}".format(accuracy_score(train_data.class_environment_score_quarter_change_pure, predicted_categories_train)))
```

```python
# Calculate accuracy
mat = confusion_matrix(valid_data.class_environment_score_quarter_change_pure, predicted_categories_valid)
print("The valid accuracy is {}".format(accuracy_score(valid_data.class_environment_score_quarter_change_pure, predicted_categories_valid)))
```

```python
# Calculate accuracy
mat = confusion_matrix(test_data.class_environment_score_quarter_change_pure, predicted_categories_test)
print("The test accuracy is {}".format(accuracy_score(test_data.class_environment_score_quarter_change_pure, predicted_categories_test)))
```

### Common Class

```python
train_common_class = train_data.class_environment_score_quarter_change_pure.sum()/len(train_data.class_environment_score_quarter_change_pure)
print("The train common class accuracy is:", train_common_class)
```

```python
valid_common_class = valid_data.class_environment_score_quarter_change_pure.sum()/len(valid_data.class_environment_score_quarter_change_pure)
print("The valid common class accuracy is:", valid_common_class)
```

```python
test_common_class = test_data.class_environment_score_quarter_change_pure.sum()/len(test_data.class_environment_score_quarter_change_pure)
print("The test common class accuracy is:", test_common_class)
```
